```python
# dataframe good
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import animation
from IPython.display import display

plt.rcParams['animation.html'] = 'jshtml'

# Simulate a DataFrame with 'Date' and 'temperature' columns
dates = pd.date_range(start='1/1/2022', periods=365)
temperatures = pd.Series(range(365)).apply(lambda x: x % 25)  # Replace with your temperature data
df = pd.DataFrame({'Date': dates, 'Temperature': temperatures})

fig, ax = plt.subplots()

# Set up the initial plot
dots = ax.scatter([], [], s=200, c='black')
ax.set_xlim(df['Date'].min(), df['Date'].max())
ax.set_ylim(df['Temperature'].min(), df['Temperature'].max())
ax.set_xlabel('Date')
ax.set_ylabel('Temperature')
ax.scatter(df['Date'], df['Temperature'], s=50, alpha=0.3)

def update(frame):
    global df, dots
    
    
    # Get the data for the current frame
    current_data = df.iloc[frame]
    current_date = current_data['Date']
    current_temperature = current_data['Temperature']
    
    # Update the scatter plot with the current data
    dots.set_offsets((current_date, current_temperature))
   
    
    return dots,

ani = animation.FuncAnimation(fig, update, frames=len(df), interval=40, blit=True)

# Display the animation using IPython's display function
display(ani)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import animation
from IPython.display import display

plt.rcParams['animation.html'] = 'jshtml'

fig, ax = plt.subplots()

# Generate random data points
np.random.seed(0)
x = np.random.rand(100)
y = np.random.rand(100)

# Set up the initial plot
dots1 = ax.scatter([], [], s=50, c='blue')
dots2 = ax.scatter([], [], s=50, c='red')
dots3 = ax.scatter([], [], s=50, c='green')
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)

def update(frame):
    # Update dots1
    dots1.set_offsets(np.c_[x[:frame], y[:frame]])
    
    # Update dots2
    dots2.set_offsets(np.c_[x[frame:2*frame], y[frame:2*frame]])
    
    # Update dots3
    dots3.set_offsets(np.c_[x[2*frame:3*frame], y[2*frame:3*frame]])
    
    return dots1, dots2, dots3

ani = animation.FuncAnimation(fig, update, frames=len(x)//3, interval=200, blit=True)

# Display the animation using IPython's display function
display(ani)
```
